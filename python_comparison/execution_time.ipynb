```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sample_size=[100,400,1000,4000,8000]
runtimes_python=[1.242521047592163,15.477382898330688,94.91009283065796,1491.2436559200287,5938.582994937897]
runtimes_c=[1.300710, 3.921253, 6.310103, 50.992920, 170.741057]
runtimes_c_opt=[1.285018, 3.606504, 5.162137, 15.003660, 43.617718]
plt.plot(sample_size,runtimes_python,'o-',label='python')
plt.plot(sample_size,runtimes_c,'o-',label='c')
plt.xlabel('M (sample size)')
plt.ylabel('Runtime (s)')
plt.title('Comparison of serial runtime for t-SNE with python and C')
plt.legend()
plt.show()
```

```python
speed_up=[runtimes_python[i]/runtimes_c_opt[i] for i in range(5)]
plt.plot(sample_size,speed_up,'o-')
plt.xlabel('M (sample size)')
plt.ylabel('Speed up')
plt.title('Speed up of parallel C tSNE compared to python implementation')
plt.show()
```
