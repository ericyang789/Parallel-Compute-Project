```python
def read_data(infile):
    df = pd.read_table(infile)    # assume a well-formatted Pandas table, tab-delimited, col headers
    ctype = list(df['type'])      # pull column 'type' out, the cell type labels
    df  = df.drop('type', axis=1)  # delete that column, leaving just the count data columns
    X   = np.log2( 1 + df.values)  # convert to log2(counts + 1). The +1 is arbitrary to avoid log(0)
    M,N = X.shape
    return X, ctype, M, N

Xs, Cs, Ms, Ns = read_data("http://mcb112.org/w13/w13-data-small.tbl")
X,  C,  M,  N  = read_data("http://mcb112.org/w13/w13-data-large.tbl")
```

### 2. implement t-SNE for yourself

Let's start laying down the pieces of a t-SNE implementation. The heart of it is two pieces: the calculation of $P$, the stochastic neighbor distribution in the original n-dimensional space, and the calculation of $Q$, in the embedded 2D space. Start with $P$. The key to it is the calculation of conditional $p_{j \mid i}$ conditional probabilities according to t-SNE's definition. We write this function assuming $\sigma_i$ values are known. Later we'll deal with optimizing $\sigma_i$ to achieve a specified "perplexity".

Here I'm going to assume that I've precalculated $\lVert x_i - x_j \rVert$ for all $i,j$ in a distance matrix $D$. I think precalculating the distance matrix will save us some compute time, since we're going to need to do these calculations a lot when we optimize for $\sigma$. 

```python
def calc_pji(D, sigmas):
    M   = D.shape[0]           # M = number of rows, samples
    pji = np.zeros((M,M))
    for i in range(M):
        for j in range(M):     # it's not symmetric, because of sigma_i
            pji[i,j] = np.exp(-D[i,j]**2 / (2 * sigmas[i]**2)) if j != i else 0.0
        Z = np.sum(pji[i])
        pji[i] = pji[i] / Z
    return pji    
```

We need that distance matrix, so a function for that, starting from our log count data $X$:

```python
def calc_D(X):
    M = X.shape[0]       # ncells
    D = np.zeros((M,M))
    for i in range(M):
        for j in range(0,i):
            D[i,j] = np.linalg.norm(X[i] - X[j])
            D[j,i] = D[i,j]
    return D
```

I'll go ahead and calculate my two distance matrices now, since I'm going to use them in some illustrations. This takes some time.

```python
#Ds = calc_D(Xs)
D  = calc_D(X)
```

We can't get $p_{j \mid i}$ calculated until we know $\sigma_i$'s, so now let's deal with that. For each $i$, we're going to fit its $\sigma_i$ separately, by dialing its value in to achieve a specified "perplexity". 

Starting from a guess at $\sigma_i$, we'll calculate the Shannon entropy $H_i$ of the current $p_{j \mid i}$ distribution; $2^{H_i}$ is the perplexity of the distribution, the effective number of neighbors.

When $\sigma_i$ is smaller, the stochastic neighbor distribution is tighter, reaching out to fewer neighbors; the perplexity is low. For larger $\sigma_i$, more neighbors, larger perplexity. Perplexity is a monotonic increasing function of $\sigma_i$, and $\sigma_i > 0$. There's got to be a solution. 

If we cast this problem in terms of an objective function $f(\sigma_i)$ that's the difference between our current perplexity and the target perplexity, then we're looking for the value of $\sigma_i$ that achieves $f(\sigma_i) = 0$. Finding scalar $x$ such that $f(x) = 0$ is a general problem called a _root-finding problem_, which arises in many sorts of situations, and tends to get a chapter of its own in books on scientific numerical methods (such as *Numerical Recipes in C*, my go-to source). Let's define that objective function. I'll define it in a particular form that's compatible with the [scipy root finding routines](https://docs.scipy.org/doc/scipy-0.14.0/reference/optimize.html):

```python
def calc_perplexity_diff(sigmai, Di, target_perplexity):   # <Di> is one row of the distance matrix.
    pji = np.exp( -Di**2 / (2 * sigmai**2))
    Z   = np.sum(pji)
    pji = pji / Z

    H = 0                     # the Shannon entropy calculation for the p_j|i given this sigma_i value
    for j in range(len(Di)):
        if pji[j] > 0:
            H -= pji[j] * np.log2(pji[j])

    perplexity = 2**H
    return perplexity - target_perplexity
```

Let's just have a look at how the calculated perplexity scales as a function of $\sigma_i$. Let's plot the perplexity for row 0 of the big data set, as we vary $\sigma_i$. 

```python
w          = 0   # which sigma_i to look at
nsamples   = 100
test_sigma = np.linspace(0.1, 4, num=nsamples)
pdiff      = [ calc_perplexity_diff(test_sigma[i], D[w], 0.) for i in range(nsamples) ]

fig2, ax2  = plt.subplots(1,1, figsize=(6,3))
ax2.plot(test_sigma, pdiff)
ax2.set_xlabel('$\sigma_i$')
ax2.set_ylabel('perplexity')
```

OK, so we need a function that dials all the $\sigma_i$ values in.

We do this by a _bisection search_ , a reliable brute force method for solving one-dimensional rootfinding problems. First we need to identify two $\sigma_i$ values $a,b$ that _bracket_ the root $f(\sigma_i)=0$, which means that $f(a)$ and $f(b)$ have opposite signs. In our problem, with $f(\sigma_i)$ a monotonically increasing function, for $a<b$ we need to identify points such that $f(a) < 0$ and $f(b) > 0$. In a function as well behaved as what we've got here, and since we know $\sigma_i > 0$, we can lay down a test point anywhere and keep moving it toward (or away) from zero until we get the sign we need. 

Having bracketed our root, we can pass the problem to the [scipy.optimize.bisect()](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.optimize.bisect.html) root finder... though bisection search is pretty easy to write, if we wanted to.

```python
def calc_sigmas(D, target_perplexity):
    M     = D.shape[0]
    sigma = np.zeros(M)
    for i in range(M):
        a, b = 1.0, 1.0   # Start testing a,b at 1.0
        while calc_perplexity_diff(a, D[i], target_perplexity) >= 0: a /= 2    # Move a in 0.5x steps until f(a) < 0
        while calc_perplexity_diff(b, D[i], target_perplexity) <= 0: b *= 2    #  ... b in 2x steps until f(a) > 0
        sigma[i] = scipy.optimize.bisect(calc_perplexity_diff, a, b, args=(D[i],target_perplexity))
    return sigma
```

At this point we've got the pieces we need to calculate our $p_{j \mid i}$. Stochastic neighbor embedding (plain SNE) would stop there, but t-SNE symmetricizes $P$, calculating quasi-joint probabilities $p_{ij}$ from the $p_{j \mid i}$. Let's now write one wrapper function that takes the input data `X` and a given target perplexity, and returns the t-SNE $P$, by calling the routines we just wrote (to calculate $p_{j \mid i}$'s), then symmetricizing them according to the t-SNE algorithm:

```python
def calc_P(X, target_perplexity):
    D      = calc_D(X)
    sigmas = calc_sigmas(D, target_perplexity)
    pji    = calc_pji(D, sigmas)
    P      = (pji + pji.T) / (2 * X.shape[0])
    return P
```

So much for $P$, the stochastic neighbor distribution in the original n-dim space. Now for $Q$. Given new points $y_i$ in the embedded 2D t-SNE space, t-SNE says to calculate $q_{ij}$ from the distance between points $y_i$ and $y_j$ according to a Student t distribution with one degree of freedom -- a Cauchy distribution:

```python
def calc_Q(Y):
    M = Y.shape[0]
    Q = np.zeros((M,M))
    Z = 0.
    for i in range(M):
        for j in range(i):  # This nested for loop leaves q_ii at zero.
            Q[i,j] = 1.0 / (1 + np.linalg.norm( Y[i] - Y[j] )**2)   # the t distribution, or Cauchy
            Q[j,i] = Q[i,j]
            Z += 2*Q[i,j]
    Q = Q / Z
    return Q
```

We're almost there! Those are the pieces we need to calculate $P$ from the input data, and $Q$ for a _given_ set of points $Y$. The optimization problem is to find an optimal set of points $Y$ that minimizes the Kullback-Leibler distance between $P$ and $Q$. We write the KL distance in a form suitable for passing it as an objective function to the [SciPy minimizer](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize).

It doesn't seem to be super well documented, but you can optionally have your objective function return the gradient too. This can save some calculation, compared to having a separate gradient-calculating function, because the calculation of $F(Y)$ and $F'(Y)$ often share computationally intensive calculations -- like here, where we're really rather just calculate $Q$ once, for a given $Y$. 

From our perspective, $Y$ is an $M \times 2$ matrix: $M$ points $y_i$, each of which is a 2D point with x/y coords in the embedded t-SNE space. Likewise, the gradient is a $M \times 2$ matrix: how much $F(Y)$ changes, for a small change in one of the two dimensional coordinates of one point $y_i$. The SciPy optimizer is very general - it doesn't want to know about the dimensionality of your problem, it just wants to deal in terms of one big vector of "current point x", and the gradient is also going to be one big vector. We use `np.flatten()` to convert a matrix to a single vector, and we use `np.reshape()` to convert a vector to a matrix.

```python
def KL_dist(Y, P):
    M      = P.shape[0]
    L      = len(Y) // M
    Y      = np.reshape(Y, (M,L))    # scipy treats Y is a vector; reshape it to an Mx2 matrix
    Q      = calc_Q(Y)
    kldist = 0.
    grad   = np.zeros((M, L))        # from our perspective, the gradient is an Mx2 matrix

    # The objective function is KL(P||Q)
    for i in range(M):
        for j in range(M):
            if P[i,j] > 0:
                kldist += P[i,j] * np.log( P[i,j] / Q[i,j] )

    # The gradient is defined on p. 2584 of van der Maaten (2008);
    # the derivation is in their appendix.
    for i in range(M):
        for j in range(M):
            grad[i] += (P[i,j] - Q[i,j]) * (Y[i] - Y[j]) * (1.0 / (1 + np.linalg.norm( Y[i] - Y[j] )**2))
        grad[i] *= 4.0

    return kldist, grad.flatten()    # .flatten(), because SciPy wants to see that gradient just as a vector.
```

Now all we have to do is choose a random starting point, and run the SciPy minimizer. t-SNE's objective function isn't convex - there are lots of local optima - so we may want to run it multiple times. Let's write a function wrapper.

```python
def my_tsne(X, target_perplexity):
    P = calc_P(X, target_perplexity)
    Y = np.random.normal(0., 1e-4, (X.shape[0], 2))   # random initial Y is drawn Normal(0, 1e-4 I)
    
    result = scipy.optimize.minimize(KL_dist, Y.flatten(), args=(P), jac=True)
    # Explication of what we pass to the minimizer:
    #  KL_dist is the objective function.
    #  Y.flatten() is our initial point Y, flattened to a single vector for SciPy's consumption.
    #  args=(P) is the interface for handing SciPy a bundle of constant additional data that it needs to pass to the
    #           objective function; here, the stochastic neighbor distribution in the original space that we're 
    #           trying to match
    # jac=True is the poorly-documented way of telling SciPy that our objective function is going to return both
    #          the f(Y) value and the f'(Y) gradient. "jac" means Jacobian matrix. The other way to do this
    #          is to pass the name of a function that calculates the gradient, i.e. jac=my_gradient_func
    # 
    # When the minimizer returns, it passes a tuple of information back. 
    # The parts we care about are:
    #    result.x        The optimized solution -- our embedded Y coordinates, as a Mx2-long vector that we'll reshape.
    #    result.success  True if the minimizer succeeded. We ought to check this.
    #    result.fun      The objective function value at the solution. 
    #    result.nit      Number of iterations it took

    Y = result.x.reshape(X.shape[0], 2)
    print (result.success)
    print (result.nit)
    
    return Y, result.fun   # Return the Y points embedded by t-SNE; and the KL distance. 
```

And at last - let's run our t-SNE on Moriarty's small data set! I'm going to set the target perplexity to a fairly small number, since both the clusters and the total number of points is so small. 

```python
import time
```

```python

def pca(X=np.array([]), no_dims=50):
    """
        Runs PCA on the NxD array X in order to reduce its dimensionality to
        no_dims dimensions.
    """

    print("Preprocessing the data using PCA...")
    (n, d) = X.shape

    X = X - np.tile(np.mean(X, 0), (n, 1))
    
    (l, M) = np.linalg.eig(np.dot(X.T, X))
    Y = np.dot(np.dot(X, M[:, 0:no_dims]),M[:, 0:no_dims].T)
    return Y

```

```python
start_time = time.time()
X = idx2numpy.convert_from_file('../data/t10k-images.idx3-ubyte')[0:2500,:,:]
M, h, w = X.shape
N = h*w
X = X.reshape((M,N))
X = X/255.0
X = pca(X, 50).real
target_perplexity = 5.0
Y, KLdist = my_tsne(X, target_perplexity)
print("--- %s seconds ---" % (time.time() - start_time))
KLdist
```

### 3. using the canned t-SNE from scikit-learn

Let's just set the scikit-learn t-SNE running on both data sets, for the asked-for range of four different perplexities... this will take some time.

```python

target_perplexities = [ 2, 5, 10, 30, 100]

Ys = []
Y  = []
for ppx in target_perplexities:
    Ys.append( TSNE(perplexity=ppx).fit_transform(Xs) )
    Y.append( TSNE(perplexity=ppx).fit_transform(X) )
```

```python
fig4, axarr = plt.subplots(5, 2, figsize=(8,14))
for k, x in enumerate(target_perplexities):
    for i in range(Ys[k].shape[0]):    # Plots on the left are the small Moriarty data set
        axarr[k,0].plot(Ys[k][i,0], Ys[k][i,1], 'o', markersize=4, mfc='w', mec=Color[Cs[i]%16] )
    axarr[k,0].set_xlabel('tSNE 1')
    axarr[k,0].set_ylabel('tSNE 2')
    axarr[k,0].set_title('small set, perplexity = {}'.format(x))
        
    for i in range(Y[k].shape[0]):     # ... on the right are the large Moriarty data set
        axarr[k,1].plot(Y[k][i,0], Y[k][i,1], 'o', markersize=4, mfc='w', mec=Color[C[i]%16] )
    axarr[k,1].set_xlabel('tSNE 1')
    axarr[k,1].set_ylabel('tSNE 2')
    axarr[k,1].set_title('large set, perplexity = {}'.format(x))       

fig4.tight_layout()
```

You can see how different choices of the perplexity affect the answer. At least in this case, a perplexity around the cluster size or smaller tends to be ok. 

It's cool that t-SNE lays out the 8 (small data set) or 16 (large data set) true clusters in an orderly fashion in 2D. The 16 corners of Moriarty's 4D hypercube in his large data set aren't too far off being a 4x4 grid in t-SNE, for perplexities of 5, 10, or 30.
